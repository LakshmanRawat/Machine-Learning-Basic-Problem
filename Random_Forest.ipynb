```python
#import libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
# Read dataset
vine = pd.read_csv('winequality-white.csv',';')
```

```python
vine.isnull().sum()
```

```python
#adding new feature as taste
vine['Taste']  = 'good'
```

```python
vine.head()
```

```python
#labeling/classifing wine quality data as good(quality>7), normal(5<quality<7) and bad(quality<5)
vine.loc[(vine.quality<5),'Taste'] = 'Bad'
vine.loc[(vine.quality>7),'Taste'] = 'Good'
vine.loc[(vine.quality>=5) & (vine.quality<=7),'Taste'] = 'Normal'
```

```python
vine.head()
```

```python
vine.tail()
```

```python
vine[vine.quality>7]
```

```python
#inititating random forest object
model = RandomForestClassifier()
```

```python
X = vine.loc[:,:'alcohol']
```

```python
X.head()
```

```python
y = vine.Taste
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# fittting model
model.fit(X_train,y_train)
```

```python
# model evalution
y_predict = model.predict(X_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```
